## F-Filtering Data

```python
# Menampilkan daftar lagu yang populer di indonesia

dataset[ dataset['country'] == 'indonesia']
```

```python
# Menampilkan daftar lagu di atas tahun 2015

dataset[ dataset['year'] > 2015]
```

```python
# Menampilkan daftar lagu yang populer di indonesia dimana lagunya di atas 2015 dan genrenya adalah 'indonesian pop'

dataset[
    (dataset['country']=='indonesia') &
    (dataset['year']>2015) &
    (dataset['genre']=='indonesian pop')
]
```

```python
# Menampilkan daftar lagu yang populer di indonesia dimana lagunya di atas 2015 dan genrenya adalah 'indonesian pop'
# dan diurutkan berdasarkan durasi terbesar ke terkecil

dataset[
    (dataset['country']=='indonesia') &
    (dataset['year']>2015) &
    (dataset['genre']=='indonesian pop')
    ].sort_values('dur', ascending=False)
```

## G-Group By

Digrup.

```python
# Mencari nilai rata-rata durasi lagu per country
# Cara: groupby berdasarkan ('country') dan data yang ditampilkan adalah['dur']

dataset

dataset_duration = dataset.groupby('country')['dur'].mean() 
dataset_duration
```

```python
dataset_min_duration_per_country = dataset.groupby('country')['dur'].min()
dataset_min_duration_per_country
```

```python
# Mencari nilai max durasi lagu per country dan kemudian diurutkan
dataset_max_duration_per_country = dataset.groupby('country')['dur'].max().sort_values()
dataset_max_duration_per_country
```

```python
# Mengubah ke dataframe

dataset_duration = dataset_duration.to_frame().reset_index()
dataset_duration.columns = ['country','duration']
dataset_duration
```

```python
dataset_duration.plot(x='country',
                   y='duration',
                   title='tingkat durasi (rata-rata)',
                   figsize=(15,5),                      # ukuran kanvas w=15 x h=5, satuan punya matplotlib (bisa langsung diimport dari pandas tanpa matplotlib)
                   kind='bar')
```
